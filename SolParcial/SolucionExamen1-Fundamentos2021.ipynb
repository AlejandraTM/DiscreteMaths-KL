## üìù Punto 3:

Consideremos el siguiente conjunto X = {a, b, c, d}. Defina una relaci√≥n tal que cumpla las siguientes propiedades:

```python
#Producto cartesiano de X con X
X=['a','b','c','d']
CX=[(i,j) for i in X for j in X]
print(set(CX))
```

1. Reflexiva, sim√©trica y no transitiva

```python
R=[(i,j) for i in X for j in X if i==j]
R.append((X[0],X[2]))
R.append((X[2],X[0]))
R.append((X[2],X[3]))
R.append((X[3],X[2]))
print(set(R))
```

```python
g=DiGraph(loops=True)
g.add_vertices(X)
g.add_edges(R)
g.show(figsize=[10,3])
```

2. Reflexiva, no sim√©trica y no transitiva

```python
#X=['a','b','c','d']
R_2=[(i,j) for i in X for j in X if i==j]
R_2.append((X[2],X[0]))
R_2.append((X[2],X[1]))
R_2.append((X[1],X[3]))
print(set(R_2))
```

```python
g_2=DiGraph(loops=True)
g_2.add_vertices(X)
g_2.add_edges(R_2)
g_2.show(figsize=[10,2])
```
